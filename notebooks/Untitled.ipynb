```python
import pandas as pd
corp_info = pd.read_csv('corp.csv',header=0)
corp_info['종목코드'] = 'A'+corp_info['종목코드']
corp_result=pd.read_excel('SAMPLE_2018.xlsx')
#corp_result
corp_data = corp_result[['종목코드','당기순이익']]
corp_total = pd.merge(corp_info,corp_data,how='left',on='종목코드')

corp_total['EPS'] = corp_total['당기순이익']/corp_total['상장주식수'].str.replace(',','').astype('float')
corp_total = corp_total[pd.notnull(corp_total['EPS'])]
corp_total = corp_total[pd.notnull(corp_total['현재가'])]

corp_total["PER"] = corp_total['현재가'].str.replace(',','').astype('int')/corp_total['EPS']

corp_filter=corp_total[(corp_total['PER']>12) & (corp_total["PER"]<=15)]
#corp_filter
corp_total.query('PER>=12 & PER <= 15')
corp_total.sort_values(by='PER')

#corp_total

```
